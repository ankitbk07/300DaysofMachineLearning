```python
import matplotlib.pyplot as plt
import sklearn.linear_model
import numpy as np 
import pandas as pd
```

```python
# load Data

oced_bli = pd.read_csv('datasets/oced_bli.csv', thousands=',')
gdp_per_capita= pd.read_csv('datasets/gdp_per_capita.csv',thousands=',',delimiter = '\t',encoding = 'latin1',na_values="n/a")
```

```python
def prepare_country_stats(oced_bli,gdp_per_capita):
    oced_bli = oced_bli[oced_bli["INEQUALITY"]=='TOT']
    oced_bli = oced_bli.pivot(index='Country',columns="Indicator",values = 'Value')
    gdp_per_capita.rename(columns = {"2015":"GDP per capita"},inplace=True)
#     print (gdp_per_capita)
    full_country_stats=  pd.merge(left = oced_bli,right= gdp_per_capita,
                             left_index= True,right_index=True)
    full_country_stats.sort_values(by = "GDP per capita",inplace = True)
    remove_indices = [0,1,6,8,33,34,35]
    keep_indices = list(set(range(35))- set(remove_indices))
#     print(keep_indices)
#     full_country_stats[["GDP per capita","Life satisfaction"]].iloc[keep_indices])
#     full_country_stats[["GDP per capita","Life satisfaction"]].iloc[keep_indices])
    return full_country_stats[["GDP per capita","Life satisfaction"]].iloc[keep_indices]




    
    
    
    
    
```
