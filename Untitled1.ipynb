```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = 2*np.random.rand(100,1)
y = 4+3*X+np.random.randn(100,1)
```

```python
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)),X] #add 1 in every instance and column vector
theta_hat = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_hat
```

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_predict = X_new_b.dot(theta_hat)
y_predict
```

```python
plt.plot(X_new,y_predict,'r-')
plt.plot(X,y,"b.")
plt.axis([0,2,0,15])
plt.show()
```

# Learning Curves 

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
def plot_learning_curves(model,X,y):
    X_train,X_val,y_train,y_val = train_test_split(X,y,test_size = 0.2)
    train_errors,val_errors = [],[]
    for m in range(1,len(X_train)+1):
        model.fit(X_train[:m],y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m],y_train_predict))
        val_errors.append(mean_squared_error(y_val,y_val_predict))
    plt.plot(np.sqrt(train_errors),"r-+",linewidth=1,label="train")
    plt.plot(np.sqrt(val_errors),'b-',linewidth=1,label ='val')
```

```python
lin_reg = LinearRegression()
plot_learning_curves(lin_reg,X,y)
```

```python
from sklearn.pipeline import Pipeline
```

```python
polynomial_regression = Pipeline([
    ("poly_features",PolynomialFeatures(degree =10,include_bias = False)),
    ("lin_ref",LinearRegression()),
])
```

```python
plot_learning_curves(polynomial_regression,X,y)
```
